```python
import numpy as np
import pandas as pd

DATA_PATH = "../Data/Vaihingen"

import os
for dirname, _, filenames in os.walk(DATA_PATH):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv(DATA_PATH + '/area2_cov_multi.csv', header=None)
```

```python
data[['x', 'y', 'z', 'cl', 'cs', 'cp', 'class']] = data[0].str.split(' ', expand=True)
```

```python
data.columns
```

```python
data.head()
```

```python
data = data.drop([0], axis=1)
```

```python
data['x'] = data['x'].astype(float)
data['y'] = data['y'].astype(float)
data['z'] = data['z'].astype(float)
data['cl'] = data['cl'].astype(float)
data['cs'] = data['cs'].astype(float)
data['cp'] = data['cp'].astype(float)
data['class'] = data['class'].astype(int)
```

```python
data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
```

```python
data= data.dropna()
```

```python
X = data.drop('class', axis=1)
y = data['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true=y_test, y_pred=y_pred)
```
