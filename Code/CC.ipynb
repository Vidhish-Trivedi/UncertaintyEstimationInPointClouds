```python
import pandas as pd

def print_null_percentage(df):
    total_rows = len(df)
    for column in df.columns:
        null_count = df[column].isnull().sum()
        percentage_null = (null_count / total_rows) * 100
        print(f"Column '{column}': {percentage_null:.2f}% null values")

file_path = '../CloudCompare/area2_cov_multi-Cloud_nbd-005.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Print the percentage of null values in each column
print_null_percentage(df)
```
