```python
from jakteristics import las_utils, compute_features, FEATURE_NAMES
import pandas as pd
import warnings
import plyfile

warnings.filterwarnings("ignore")
```

```python
pickle_file = "C:\\Users\\Vidhish17\\Desktop\\20Credit\\data\\paris-lille-10\\training_10_classes\\Paris.ply"
selected_features = ['eigenvalue_sum', 'omnivariance', 'eigenentropy', 'anisotropy', 'planarity', 'linearity', 'PCA1', 'PCA2', 'surface_variation', 'sphericity', 'verticality', 'nx', 'ny', 'nz']
```

```python
print(FEATURE_NAMES)
```

```python
points = plyfile.PlyData.read(pickle_file)
```

```python
data = points.elements[0].data.tolist()
```

```python
df = pd.DataFrame([(row[0], row[1], row[2], row[4]) for row in data], columns=['x', 'y', 'z', 'class'])
```

```python
df.describe()
```

```python
classes = df['class'].unique()
classes
```

```python
xyz = df.drop(columns=['class']).astype("float").values
```

```python
xyz
```

```python
%%time

features = compute_features(xyz, search_radius=0.25, feature_names=selected_features)
```

```python
features = pd.DataFrame(features, columns=selected_features)
```

```python
points_with_features = pd.concat([df, features], axis=1)
points_with_features
```
